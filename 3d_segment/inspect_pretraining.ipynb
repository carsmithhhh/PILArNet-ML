```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('train_loss_no_norm.txt', header=None, names=['epoch', 'it', 'train_loss', 'cosine_sim'])

plt.plot(df.index, np.log(df['train_loss']), label='Log Train Loss')
plt.xlabel('Iteration')
plt.ylabel('Train Loss')
plt.title('SimCLR Pretraining Loss')
plt.legend()
plt.show()
```

```python
plt.plot(df.index, np.log(df['cosine_sim']), label='Cosine sim')
plt.xlabel('Iteration')
plt.ylabel('Cosine sim')
plt.title('SimCLR Positive Pair Cosine Sim')
plt.legend()
plt.show()
```
