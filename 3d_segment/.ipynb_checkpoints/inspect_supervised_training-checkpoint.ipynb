```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('supervised_results.txt', header=None, names=['epoch', 'it', 'train_loss', 'train_acc'])
df2 = pd.read_csv('supervised_val_stats.txt', header=None, names=['epoch', 'it', 'val_loss', 'val_acc'])

plt.plot(df.index, df['train_loss'], label='Train Loss')
plt.scatter(np.array([i * 140 for i in range(1, 6)]), df2['val_loss'], label='Val Loss', color='red')
plt.xlabel('Iteration')
plt.ylabel('Train Loss')
plt.title('Supervised Full UNet Cross-Entropy Loss')
plt.legend()
plt.show()
```

```python
plt.plot(df.index, df['train_acc'], label='Train Acc')
plt.scatter(np.array([i * 140 for i in range(1, 6)]), df2['val_acc'], label='Val Acc', color='red')
plt.xlabel('Iteration')
plt.ylabel('Train Accuracy')
plt.title('Supervised Full UNet Accuracy')
plt.legend()
plt.show()
```

```python
df = pd.read_csv('supervised_val_stats.txt', header=None, names=['epoch', 'it', 'val_loss', 'val_acc'])

plt.plot(df.index, df['val_loss'], label='Val Loss')
plt.xlabel('Iteration')
plt.ylabel('Val Loss')
plt.title('Supervised Full UNet Cross-Entropy Validation Loss')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npy file
data = np.load('supervised_results_new.npy', allow_pickle=True).item()  # .item() if it's a saved dict

print(data.keys())
# Expecting the structure to be something like:
# data = {
#     "train_loss": [...],
#     "train_acc": [...],
#     "val_loss": [...],
#     "val_acc": [...],
#     "iters": [...],         # optional, if iteration numbers are stored
#     "val_iters": [...]      # optional, or use e.g. every Nth from "iters"
# }

train_loss = data['train_loss']
val_loss = data['eval_loss']

# Optional: use stored iteration numbers if available
iters = data.get('iters', np.arange(len(train_loss)))
iters = np.arange(len(data['train_loss']))
# val_iters = data.get('val_iters', np.array([i * 140 for i in range(1, len(val_loss) + 1)]))

# Plotting
plt.plot(iters, train_loss, label='Train Loss')
plt.scatter(iters, val_loss, label='Val Loss', color='red')
plt.xlabel('Iteration')
plt.ylabel('Cross-Entropy Loss')
plt.title('Supervised Full UNet Cross-Entropy Loss')
plt.legend()
plt.grid(True)
plt.show()
```
